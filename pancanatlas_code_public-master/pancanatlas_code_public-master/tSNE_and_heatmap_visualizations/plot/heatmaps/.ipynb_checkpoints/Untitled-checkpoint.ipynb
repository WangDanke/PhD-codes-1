```python
import os
import sys

sys.path.append('../..')
import config
import utils

sys.path.append('..')
import plot_utils

import numpy as np
import pandas as pd

import seaborn as sns
import scipy.spatial as sp
import scipy.cluster.hierarchy as hc

import rep_dists as PKG;
```

```python

path = os.path.join(config.embed_dir, 'expression', 'data.tsv')
outdir = os.path.join(config.plot_dir, 'expression', 'heatmaps')
desc = 'Expression'
try:
    df = utils.load_large_df(path.replace('.tsv', ''))
except IOError:
    df = pd.read_csv(path, sep='\t', index_col=0)

keep_cols = PKG.filter_to_high_var(df.values, df.columns, 5000)
df = df.iloc[:, keep_cols]
metadata_df = utils.load_metadata_df(config.metadata_path, df.index)
medians = PKG.collapse_to_median(df, metadata_df['cnc'])
cnc_to_color = utils.load_color_scheme(config.color_scheme_path)
```

```python
fig, ax = plt.subplots()
ax.imshow(medians.values.astype(float))
ax.set_aspect(50)
plt.show()
```

```python
import scipy.spatial.distance as ssd
import scipy.cluster.hierarchy as sch
dists = ssd.pdist(medians)
linkage = sch.linkage(dists)

```

```python
import seaborn as sns
ordered_medians = medians.iloc[:, medians.median().argsort().values].astype(float)
sns.clustermap(ordered_medians, col_cluster=False)
```
